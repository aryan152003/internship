```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\internship\job_descriptions.csv')
```

```python
df.head(5)
```

```python
asian_countries = [
    "Afghanistan", "Armenia", "Azerbaijan", "Bahrain", "Bangladesh", "Bhutan", 
    "Brunei", "Cambodia", "Georgia", "India", "Indonesia", "Iran", "Iraq", 
    "Israel", "Japan", "Jordan", "Kazakhstan", "Kuwait", "Kyrgyzstan", "Laos", 
    "Lebanon", "Malaysia", "Maldives", "Mongolia", "Myanmar", "Nepal", 
    "North Korea", "Oman", "Pakistan", "Palestine", "Philippines", "Qatar", 
    "Saudi Arabia", "Singapore", "South Korea", "Sri Lanka", "Syria", 
    "Tajikistan", "Thailand", "Timor-Leste", "Turkey", "Turkmenistan", 
    "United Arab Emirates", "Uzbekistan", "Vietnam", "Yemen"]
df = df[~df['Country'].isin(asian_countries)]

df = df[~df['Country'].str.startswith('C')]

df = df[df['latitude'] >= 10]

df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-06-01')

df = df[(df['Job Posting Date'] >= start_date) & (df['Job Posting Date'] <= end_date)]

df = df[df['Qualifications'] == 'B.Tech']

df = df[df['Preference'] == 'Female']

df = df[(df['Role'] == 'Data Analyst') & (df['Job Title'] == 'Data Scientist')]
```

```python
if df.empty or 'Company' not in df.columns or df['Company'].isnull().all():
    print("Their is no data with these filters in the dataset.")
else:
    company_counts = df['Company'].value_counts().head(10)

    # Plotting
    plt.figure(figsize=(10, 6))
    company_counts.plot(kind='bar', color='skyblue')
    plt.title('Top 10 Companies with Max Data Analyst Roles and Data Scientist Job Titles (Female Preference)')
    plt.xlabel('Company Name')
    plt.ylabel('Number of Postings')
    plt.xticks(rotation=45)
    plt.show()
```
