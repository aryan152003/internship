```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv(r'D:\internship\job_descriptions.csv')
```

```python
df.head(5)
```

```python
african_countries = [
    "Algeria", "Angola", "Benin", "Botswana", "Burkina Faso", "Burundi", 
    "Cabo Verde", "Cameroon", "Central African Republic", "Chad", "Comoros", 
    "Congo", "Democratic Republic of the Congo", "Djibouti", "Egypt", 
    "Equatorial Guinea", "Eritrea", "Eswatini", "Ethiopia", "Gabon", 
    "Gambia", "Ghana", "Guinea", "Guinea-Bissau", "Ivory Coast", "Kenya", 
    "Lesotho", "Liberia", "Libya", "Madagascar", "Malawi", "Mali", 
    "Mauritania", "Mauritius", "Morocco", "Mozambique", "Namibia", "Niger", 
    "Nigeria", "Rwanda", "Sao Tome and Principe", "Senegal", "Seychelles", 
    "Sierra Leone", "Somalia", "South Africa", "South Sudan", "Sudan", 
    "Tanzania", "Togo", "Tunisia", "Uganda", "Zambia", "Zimbabwe"
]
df = df[~df['Country'].isin(african_countries)]
filtered_df = df[
    (df['Qualifications'].isin(['B.Tech', 'M.Tech', 'PhD'])) &
    (df['Work Type'] == 'Full time') &
    (df['Job Title'].str.startswith('D')) &
    (df['Preference'] == 'Male') &
    (df['Company Size'] > 80000) &
    (df['Contact'].str.startswith('A')) &
    (df['Job Portal'] == 'Indeed')
]
```

```python
if not filtered_df.empty:
    company_job_count = filtered_df['Company'].value_counts().head(10)

    plt.figure(figsize=(10, 6))
    company_job_count.plot(kind='bar', color='orange')
    plt.title('Top Companies by Job Title Starting with "D" (Male Preference)')
    plt.xlabel('Company')
    plt.ylabel('Number of Job Postings')
    plt.xticks(rotation=45)
    plt.show()
    m = folium.Map(location=[0, 0], zoom_start=2)
    for idx, row in filtered_df.iterrows():
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=f"Company: {row['company_name']}\nJob Title: {row['job_title']}\nLocation: {row['latitude']}, {row['longitude']}",
            tooltip="Click to see the location"
        ).add_to(m)
    m.save('job_map.html')
else:
    print("The filtered DataFrame is empty. No data to display.")
```
