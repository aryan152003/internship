```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\internship\job_descriptions.csv')
```

```python
df.head(5)
```

```python
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])
df['Experience'] = pd.to_numeric(df['Experience'], errors='coerce')
df['Salary Range'] = pd.to_numeric(df['Salary Range'], errors='coerce')
filtered_df = df[
    (df['Country'].isin(['India', 'Germany'])) &
    (df['Qualifications'] == 'B.Tech') &
    (df['Work Type'] == 'Full time') &
    (df['Experience'] > 2) &
    (df['Job Title'] == 'Data Science') &
    (df['Salary Range'] > 100000) &
    (df['Job Portal'] == 'Indeed') &
    (df['Preference'] == 'Female') &
    (df['Job Posting Date'] < '2023-08-01')]
```

```python
country_job_count = filtered_df['Country'].value_counts()
if not country_job_count.empty:
    colors = ['orange', 'green']  
    plt.figure(figsize=(8, 6))
    country_job_count.plot(kind='bar', color=colors)
    plt.title('Data Science Jobs for B.Tech Qualified Females (Full-time, >$100k Salary)')
    plt.xlabel('Country')
    plt.ylabel('Number of Job Postings')
    plt.xticks(rotation=0)
    plt.show()
else:
    print("No job postings available for the specified criteria.")
```
